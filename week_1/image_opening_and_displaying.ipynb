# Image Opening and Displaying

## Displaying using Matplotlib

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
im = cv.imread('emma.jpg')
fig, ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()
```

## Displaying using OpenCV

```python
im = cv.imread('emma.jpg')
cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image', im)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Displaying Image Properties

```python
print('Image Shape:', im.shape)
print('Image Data Type:', im.dtype)
print('Image Size:', im.size)
```

## Increasing the Brightness

```python
im = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
im2 = cv.add(im, 100)
fig, ax = plt.subplots(1,2,figsize=(6,8))
ax[0].imshow(im, cmap="gray", vmin=0, vmax=255)
ax[1].set_title('Original Image')
ax[1].imshow(im2, cmap="gray",vmin=0,vmax=255)
ax[1].set_title('Brightness Increased')
for a in ax:
    a.axis('off')
plt.show()
```

```python
im = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
im2 = im + 100 # This differs the output from the above
fig, ax = plt.subplots(1,2,figsize=(6,8))
ax[0].imshow(im, cmap="gray", vmin=0, vmax=255)
ax[1].set_title('Original Image')
ax[1].imshow(im2, cmap="gray",vmin=0,vmax=255)
ax[1].set_title('Brightness Increased')
for a in ax:
    a.axis('off')
plt.show()
```
